```julia
using HomotopyContinuation
```

```julia
@var a[1:3,1:6] α[1:2] r x[1:3] y[1:3] λ[1:3] s[1:3]
```

```julia
inC = [-y[1]-x[1]^2+2x[1]+5+x[1]*y[1], 
        -y[2] + 2x[2]^2+5x[2]-8, 
        -y[3] + 8x[3]^2-3x[3]-2]
```

```julia
inC = [-y[1]-x[1]^2+2x[1]+5+x[1]*y[1], 
        -y[2] + 2x[2]^2+5x[2]-8, 
        -y[3] + 8x[3]^2-3x[3]-2]
```

```julia
inC = [rand_poly(Float64, [x[i],y[i]],2) for i = 1:3]
```

```julia
circle = [(x[1]-α[1])^2+(y[1]-α[2])^2 - r,
        (x[2]-α[1])^2+(y[2]-α[2])^2 - r,
        (x[3]-α[1])^2+(y[3]-α[2])^2 - r]
```

```julia
normal_space = [ [x[i] - α[1] - λ[i]*(differentiate(inC[i], x[i])), y[i] - α[2] - λ[i]*(differentiate(inC[i], y[i]))] for i = 1:3]
```

```julia
normal_space = vcat(normal_space...)
```

```julia
sys = System(vcat(inC, circle, normal_space))
```

```julia
result = solve(sys)
```

```julia
sols = solutions(result)
```

```julia
sols = map(i -> [i[1], i[8], i[9]], sols)
```

```julia
unique_points(sols)
```
