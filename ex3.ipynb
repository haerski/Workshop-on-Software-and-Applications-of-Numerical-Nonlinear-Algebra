```julia
using HomotopyContinuation
```

```julia
@var x[1:3] y[1:3] z[1:3] X Y Z s a b c λ[1:3]
```

```julia
f = rand_poly([X,Y,Z],4)
```

```julia
inX = [ evaluate(f, [X,Y,Z] => [x[i],y[i],z[i]]) for i = 1:3 ];
```

```julia
using LinearAlgebra
```

```julia
distinct = det([x y z])*s - 1
```

```julia
plane = [ a*x[i] + b*y[i] + c*z[i] + 1 for i = 1:3 ]
```

```julia
tangent = [ λ[i]*[a,b,c] - evaluate(differentiate(f, [X,Y,Z]), [X,Y,Z] => [x[i],y[i],z[i]]) for i = 1:3 ]
```

```julia
sys = System(vcat(tangent...,plane, distinct, inX))
```

```julia
solve(sys)
```
